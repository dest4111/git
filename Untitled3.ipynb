цей код малює паралелограм

```python
import matplotlib.pyplot as plt

# Координати вершин паралелограма
x = [0, 2, 3, 1]
y = [0, 1, 3, 2]

# Створення графіку та намалювання паралелограма
plt.plot(x + [x[0]], y + [y[0]], marker='o')
plt.fill(x, y, alpha=0.3)  # Заповнення паралелограма кольором з прозорістю

# Налаштування відображення
plt.xlim(-1, 4)
plt.ylim(-1, 4)
plt.grid(True)

# Відображення графіку
plt.show()
```
